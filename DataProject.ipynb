```python
#installing libraries
import pandas as pd
```

```python
trxn=pd.read_csv('Retail_Data_Transactions.csv')
trxn
```

```python
response=pd.read_csv('Retail_Data_Response.csv')
response
```

```python
df=trxn.merge(response,on='customer_id',how='left')
df
```

```python
#features
df.dtypes
df.shape
df.head()
df.tail()
```

```python
df.describe()
```

```python
#check missing values
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
df['trans_date']=pd.to_datetime(df['trans_date'])
df
```

```python
set(df['response'])
```

```python
df['response']=df['response'].astype('int64')
df
```

```python
set(df['response'])
```

```python
#check for outliers
#z-score

from scipy import stats
import numpy as np

#calculate z score
z_score=np.abs(stats.zscore(df['tran_amount']))

#set threshold 
threshold=3 #if outlier can be if it is +3 or -3
outliers=z_score>threshold
print(df[outliers])
```

```python
#calculate z score
from scipy import stats
import numpy as np
z_score=np.abs(stats.zscore(df['response']))

#set threshold 
threshold=3 #if outlier can be if it is +3 or -3
outliers=z_score>threshold
print(df[outliers])
```

```python
import  seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=df['tran_amount'])
plt.show()
```

```python

#creating new coln
df['month']=df['trans_date'].dt.month
df
```

```python
#which 3 month have had highest transtion amount
montly_sales=df.groupby('month')['tran_amount'].sum()
montly_sales=montly_sales.sort_values(ascending=False).reset_index().head(3)
montly_sales
```

```python
# coustomers  having highest number of order
coustomer_count=df['customer_id'].value_counts().reset_index()
coustomer_count
#sort
top_5_cus=coustomer_count.sort_values(by='count',ascending=False).head(5)
top_5_cus
```

```python
sns.barplot(x='customer_id',y='count',data=top_5_cus)
```

```python
# coustomers  having highest value of order

coustomer_sales=df.groupby('customer_id')['tran_amount'].sum().reset_index()
coustomer_count
#sort
top_5_sal=coustomer_sales.sort_values(by='tran_amount',ascending=False).head(5)
top_5_sal
```

```python
sns.barplot(x='customer_id',y='tran_amount',data=top_5_sal)
```
